```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erf

# 활성화 함수 정의
def relu(x):
    return np.maximum(0, x)

def gelu(x):
    return 0.5 * x * (1 + erf(x / np.sqrt(2)))

def swish(x):
    return x / (1 + np.exp(-x))

def swiglu(x):
    # 실제 SwiGLU는 입력을 두 부분으로 나누어 계산되지만,
    # 여기서는 a = b = x로 가정: f(x) = x * swish(x) = x^2 * sigmoid(x)
    return x * swish(x)

# x 값 범위 설정
x = np.linspace(-5, 5, 400)

# 각 함수 계산
y_relu = relu(x)
y_gelu = gelu(x)
y_swish = swish(x)
y_swiglu = swiglu(x)

# 그래프 그리기
plt.figure(figsize=(8, 6))
plt.plot(x, y_relu, label="ReLU", linewidth=2)
plt.plot(x, y_gelu, label="GELU", linewidth=2)
plt.plot(x, y_swish, label="Swish", linewidth=2)
plt.plot(x, y_swiglu, label="SwiGLU (a=b=x)", linewidth=2)

plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Activation Functions")
plt.legend()
plt.grid(True)
plt.show()
```
