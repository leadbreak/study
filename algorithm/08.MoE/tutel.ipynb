```python
import torch
from tutel import moe as tutel_moe
```

```python
x = torch.ones([6, 1024], device='cuda:0')
x
```

```python
# create MoE
moe_layer = tutel_moe.moe_layer(
    gate_type = {'type':'top', 'k':2},
    model_dim = x.shape[-1],
    experts = {
        'count_per_node':2,
        'type':'ffn',
        'hidden_size_per_expert':2048,
        'activation_fn': lambda x: torch.nn.functional.gelu(x),
    }
)
```

```python
# Cast to GPU
moe_layer = moe_layer.to('cuda:0')
```

```python
# Forward MoE:
y = moe_layer(x)
y
```
