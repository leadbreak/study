1. ReLU(Rectified Linear Unit):
- 정의: $\text{ReLU}(x) = \max(0, x)$
- 특징1: 음수 입력에 대해 0을 출력하며, 양수 입력에 대해서는 x를 그대로 출력. 매우 간단하고 계산 효율적
- 특징2: 비선형성이 강하고, 음수 입력을 완전히 배제해 그레디언트 소실 문제 발생

2. Swish:
- 정의: $ \text{Swish}(x) = x \cdot \sigma(\beta x) $
- 특징1: 입력에 대해 부드러운 비선형 변환을 제공하며, $\sigma$는 시그모이드 함수($ \sigma(x) = \frac{1}{1 + e^{-x}} $)를 의미
- 특징2: 입력값이 매우 크거나 작을 때 ReLU보다 정보 손실이 적으며, 음수 입력에 대해 작은 값이 활성화 돼 그레디언트 소실 문제를 완화

3. GELU:
- 정의: $\text{GELU}(x) = x \cdot \Phi(x)$
- 특징1 : 입력값을 정규분포의 누적분포함수(CDF)를 사용하여 변환(위 수식에서 '$\Phi(x)$'가 표준정규분포의 CDF)
- 특징2 : ReLU나 Swish에 비해 더 복잡한 계산을 필요로 하나, 대신 더 복잡한 문제를 해결할 수 있음

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# 정의된 범위와 분해능 설정
x = np.linspace(-5, 5, 200)

# ReLU 활성화 함수
relu = np.maximum(0, x)

# Swish 활성화 함수
beta = 1  # Swish 파라미터
swish = x / (1 + np.exp(-beta * x))

# GELU 활성화 함수
gelu = x * stats.norm.cdf(x)

# 고해상도 설정
plt.figure(figsize=(14, 8), dpi=100)

# 선 스타일 및 두께 조정
plt.plot(x, relu, label='ReLU', color='red', linestyle='-', linewidth=4)
plt.plot(x, swish, label='Swish', color='blue', linestyle='--', linewidth=4)
plt.plot(x, gelu, label='GELU', color='green', linestyle='-.', linewidth=4)

# 축 및 레이블 스타일 설정
plt.xlabel('x', fontsize=16, fontweight='bold')
plt.ylabel('f(x)', fontsize=16, fontweight='bold')
plt.title('Comparison of Activation Functions: ReLU, Swish, and GELU', fontsize=18, fontweight='bold')

# 범례 및 격자 스타일
plt.legend(fontsize=14, frameon=True)
plt.grid(True, linestyle='--', alpha=0.7)

# 축 숫자 크기 조정
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# 배경색 설정
plt.gca().set_facecolor('#EFEFEF')

plt.show()
```
